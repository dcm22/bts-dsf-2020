```
!curl -o FremontBridge.csv https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD
```

```
import pandas as pd
url = '/content/FremontBridge.csv'
bikes = pd.read_csv(url,index_col='Date', parse_dates=True)
```

```
bikes
```

```
bikes.dtypes
```

```
### bikes['Date'] = pd.to_datetime(bikes["Date"]) 
#bikes.dtypes
#bikes = bikes.set_index('Date')
#bikes.dtypes
###
```

```
##When are the busiest hours per week in Total?

bikes.loc[:,'weekday'] = bikes.index.weekday
bikes2 = bikes
#Turn 'weekday' into a category 
bikes2["weekday"] = bikes2["weekday"].astype("category")

#Rename the categories
bikes2["weekday"].cat.categories = ["Mon", "Tues", "Wed", "Thurs", "Fri", "Sat", "Sun"]

bikes2["weekday"] = bikes2["weekday"].cat.set_categories(["Mon", "Tues", "Wed", "Thurs", "Fri", "Sat", "Sun"])
bikes2
```

```
bikes3 = bikes2.groupby(bikes2.index.hour).sum()
bikes3['Fremont Bridge Total'].agg(max)
bikes3['Fremont Bridge Total'].plot()
bikes3
```

The busiest hour in total is from 17:00 to 18:00

```
#Where will you be more comfortable (alone) riding during the week in West or East?
bikes4 = bikes2.groupby(bikes2.index.hour).sum()
bikes4.plot()
```

You'd be more comfortable on the EAST side as there are less people in the way. Though from 3:00 until 14:00 it is slightly busier than the west.

```
#What about riding during the business days? When and where are the best hours to ride alone between 06:00 and 10:00? And between 15:00 and 20:00?
bikes_bday = bikes[bikes.weekday != 'Sun']
bikes_bday = bikes_bday[bikes_bday.weekday != 'Sat']
```

```
bikes_bday = bikes_bday.groupby(bikes_bday.index.hour).sum()
bikes_bday.plot()
bikes_bday
```

During business days it is also better to use the EAST sidewalk, especially between 15:00 and 20:00. Though between 6:00 and 10:00 one should use the WEST wakway

```
#What happened during the coronavirus pandemic? Did the daily riding hours change? How many riders are passing through the bridge comparing 2018 and 2020?
bikes['doy'] = bikes.index.dayofyear
bikes['Year'] = bikes.index.year
bikes['Month'] = bikes.index.month
```

```
piv_daily = pd.pivot_table(bikes, index=['doy'],columns=['Year'], values=['Fremont Bridge Total'])
piv_daily.head()
```

```
piv_monthly = pd.pivot_table(bikes, index=['Month'],columns=['Year'], values=['Fremont Bridge Total'])
piv_monthly
```

```
piv_daily.plot(figsize=(16, 12))
piv_monthly.plot(figsize=(16, 12))
```

The daily usage is down as we can see in the graphs and tables. It was hard to determine this with the daily graph, therefore I grouped the values by month to get a clearer view of the trends

```
import seaborn as sns 
piv_daily1 = piv_daily.drop(piv_daily.columns[0:6], axis=1)
piv_monthly1 = piv_monthly.drop(piv_monthly.columns[0:6], axis=1)

piv_daily2 = piv_daily1.drop(piv_daily1.columns[1], axis=1)
piv_monthly2 = piv_monthly1.drop(piv_monthly1.columns[1], axis=1)

piv_daily2.plot(figsize=(16, 12))
piv_monthly2.plot(figsize=(16, 12))

piv_daily2.head()
piv_monthly2
```

As we can see from the table and graphs, there has been a decrease in cyclists from February 2020 compared to the numbers from 2018.
